## Metrics for Coding Scheme A: explicit/implicit/relchanges/releconomy vs. irrelevant

```python
#Calculate metrics for Coding Scheme A, Not Inequality/Irrelevant category
tp1_irrel = df[(df['b1']==2) & (df['linearSVC_binary1']==2)]
tn1_irrel = df[(df['b1']==1) & (df['linearSVC_binary1']==1)]
fp1_irrel = df[(df['b1']==1) & (df['linearSVC_binary1']==2)]
fn1_irrel = df[(df['b1']==2) & (df['linearSVC_binary1']==1)]

trupos1_irrel = float(len(tp1_irrel.index))
truneg1_irrel = float(len(tn1_irrel.index))
falpos1_irrel = float(len(fp1_irrel.index))
falneg1_irrel = float(len(fn1_irrel.index))

precision_irrel = trupos1_irrel/(trupos1_irrel + falpos1_irrel)
recall_irrel = trupos1_irrel/(trupos1_irrel + falneg1_irrel)

f1_irrel = 2*((precision_irrel * recall_irrel)/(precision_irrel + recall_irrel))

print("Number of articles hand-coded as irrelevant: %d" % len(df[df['b1']==2].index))
print("Precision Irrelevant: %f" % precision_irrel)
print("Recall Irrelevant: %f" % recall_irrel)
print("F1 Irrelevant: %f" % f1_irrel)
```

```python
#Calculate overall metrics for Coding Scheme A
precision_total = (trupos1_rel + trupos1_irrel)/((trupos1_rel+trupos1_irrel) + (falpos1_rel+falpos1_irrel))
recall_total = (trupos1_rel + trupos1_irrel)/((trupos1_irrel+trupos1_rel) + (falneg1_irrel + falneg1_rel))
f1_total = 2*((precision_total * recall_total)/(precision_total + recall_total))
f1_total

print("Number of articles in test set: %d" % len(df.index))
print("Precision Total: %f" % precision_total)
print("Recall Total: %f" % recall_total)
print("F1 Total: %f" % f1_total)
```
